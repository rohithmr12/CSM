```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, diff

# Define the symbolic variable
x = symbols('x')

# Define the function and its derivative
f = x**10 + x - 10
fdash = diff(f, x)

# Convert the symbolic expressions to Python functions
f_func = lambdify(x, f, 'numpy')
fdash_func = lambdify(x, fdash, 'numpy')

# Generate x values
x_values = np.linspace(-2, 2, 400)

# Calculate y values for the function and its derivative
y_values = f_func(x_values)
y_dash_values = fdash_func(x_values)

# Plot the function and its derivative
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='f(x)')
plt.plot(x_values, y_dash_values, label="f'(x)")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title("Function and its Derivative")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```
